```python
path = ("data/EP_COMBINED_no_date_filter.csv")
df = pd.read_csv(path)
```

```python
df.shape
```

```python
df.columns
```

```python
quantile = 0.1
```

```python
col = "Gap%"
q_low = df[col].quantile(quantile)
q_hi  = df[col].quantile(1 - quantile)

df_filtered = df[(df[col] < q_hi) & (df[col] > q_low)]
```

```python
col = "bh_high_returns"
q_low = df[col].quantile(quantile)
q_hi  = df[col].quantile(1 - quantile)

df_filtered = df[(df[col] < q_hi) & (df[col] > q_low)]
```
