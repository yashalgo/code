```python
def get_estimates_df(ticker):
    x = f"""=FS_EarningsCalendar("1990-01-01", "2023-01-31", "{ticker}")"""
    while True:
        if len(request_times) >= 600 and time.time() - request_times[-600] <= 60:
            time.sleep(0.1)
            continue
        break
    request_times.append(time.time())
    sheet_instance.update('A1', x, raw=False)
    df = pd.DataFrame(sheet_instance.get_all_records())
    return df
```
