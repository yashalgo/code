```python
tickers = pd.read_csv(data_ / "US/US_STOCKS.csv")['Ticker']
```

```python
request_times = []
```

```python
files = glob('*.csv')
```

```python
len(files), len(tickers)
```

```python
for ticker in tickers:
    try:
        outfile = ticker + '.csv'
        if outfile in files:
            continue
        start_time = time.time()
        while len(request_times) >= 2000 and (time.time() - request_times[-2000]) < 3600:
            time.sleep(1)
        df2 = yf.download(ticker, start='2000-01-01')
        request_times.append(time.time())
        df2.reset_index(level=0, inplace=True)
        df2["Date"] = df2["Date"].apply(
            lambda x: datetime.strptime(datetime.strftime(x,  "%Y-%m-%d"), "%Y-%m-%d")
        )
        df2.to_csv(outfile)
        print(ticker, df2.shape, time.time() - start_time)
    except:
        print('Error for: ', ticker)
```
